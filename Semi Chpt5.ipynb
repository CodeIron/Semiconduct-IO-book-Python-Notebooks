#Section 2.2 Plot data in eV
The data are converted from wavelength (nm) to photon energy in eV and then plotted for TE and TM polarisation

#Section 2.3 Plot difference spectra

```
import matplotlib.pyplot as plt
import numpy as np
eV=list(wave_len)#convert from tuple which is immutable
for i in range(len(wave_len)):
    eV[i]=1240/ wave_len[i] #converts photon scale from wavelngth in nm to eV
baseTE_spectrum=TEM01
delta_TEM01=np.subtract(TEM01,baseTE_spectrum)
delta_TEM02=np.subtract(TEM02,baseTE_spectrum)
delta_TEM03=np.subtract(TEM03,baseTE_spectrum)
delta_TEM04=np.subtract(TEM04,baseTE_spectrum)
delta_TEM05=np.subtract(TEM05,baseTE_spectrum)
baseTM_spectrum=TMM01
delta_TMM01=np.subtract(TMM01,baseTM_spectrum)
delta_TMM02=np.subtract(TMM02,baseTM_spectrum)
delta_TMM03=np.subtract(TMM03,baseTM_spectrum)
delta_TMM04=np.subtract(TMM04,baseTM_spectrum)
delta_TMM05=np.subtract(TMM05,baseTM_spectrum)
fig,a = plt.subplots(2,sharex=True, figsize=(16, 16)) #sets size of figures
a[1].set_title('TM Reverse Bias Difference Spectra',fontsize=24) #writes title
a[0].set_title('TE Reverse Bias Difference Spectra',fontsize=24) #writes title
a[1].set_ylabel("$\Delta$ Absorption coefficient [1/cm]",fontsize=20)#writes y axis label
plt.xlabel('Photon energy [eV]',fontsize=20)#writes x axis label
a[0].set_ylabel('$\Delta$ Absorption coefficient [1/cm]',fontsize=20)#writes y axis label
a[0].plot(eV,delta_TEM01,'k-',label="Applied Field =0kV/cm") #Plots TEM01 Spectrum
a[0].plot(eV,delta_TEM02,'r-',label="Applied Field =37kV/cm") #Plots TEM02 Spectrum
a[0].plot(eV,delta_TEM03,'g-',label="Applied Field =74kV/cm") #Plots TEM03 Spectrum
a[0].plot(eV,delta_TEM04,'y-',label="Applied Field =111kV/cm") #Plots TEM04 Spectrum
a[0].plot(eV,delta_TEM05,'b-',label="Applied Field =148kV/cm") #Plots TEM05 Spectrum
a[1].plot(eV,delta_TMM01,'k-',label="Applied Field =0kV/cm") #Plots TEM01 Spectrum
a[1].plot(eV,delta_TMM02,'r-',label="Applied Field =37kV/cm") #Plots TEM02 Spectrum
a[1].plot(eV,delta_TMM03,'g-',label="Applied Field =74kV/cm") #Plots TEM03 Spectrum
a[1].plot(eV,delta_TMM04,'y-',label="Applied Field =111kV/cm") #Plots TEM04 Spectrum
a[1].plot(eV,delta_TMM05,'b-',label="Applied Field =148kV/cm") #Plots TEM05 Spectrum
a[0].legend(loc=0)#writes legend on graph
a[1].legend(loc=0)#writes legend on graph
a[0].grid()#writes grid on graph
a[1].grid()#writes grid on graph
plt.show()
```

#Section 2.4 for TE data plot extinction coefficient and calculate  and plot change in refractive index

```
from scipy import special
import matplotlib.pyplot as plt
import numpy as np
import math
import ipywidgets as widgets
import scipy.fftpack as ft # Contains a function that will compute the Hilbert transform
from matplotlib.transforms import Transform
for i in range(len(wave_len)):
    eV[i]=1240/ wave_len[i] #converts photon scale from wavelngth in nm to eV
pi=3.1415#value of pi
c=2.9979245E8#speed of light m/s
wavelength=np.asarray(wave_len)
delta_TEM05_n=np.asarray(delta_TEM05)
delta_TEM04_n=np.asarray(delta_TEM04)
delta_TEM03_n=np.asarray(delta_TEM03)
delta_TEM02_n=np.asarray(delta_TEM02)
extinction_coeff_TEM05=wavelength*1E-7*delta_TEM05_n/2/pi#converts the absorption coefficient(1/cm) to extinction coefficient see Whittaker wavelength converted from nm to cm 
efract_TEM05=ft.hilbert(extinction_coeff_TEM05)#computes the change in refractive index from change in in extinction coefficient spectrum
extinction_coeff_TEM04=wavelength*1E-7*delta_TEM04_n/2/pi#converts the absorption coefficient(1/cm) to extinction coefficient see Whittaker wavelength converted from nm to cm 
efract_TEM04=ft.hilbert(extinction_coeff_TEM04)#computes the change in refractive index from change in in extinction coefficient spectrum
extinction_coeff_TEM03=wavelength*1E-7*delta_TEM03_n/2/pi#converts the absorption coefficient(1/cm) to extinction coefficient see Whittaker wavelength converted from nm to cm 
efract_TEM03=ft.hilbert(extinction_coeff_TEM03)#computes the change in refractive index from change in extinction coefficient spectrum
extinction_coeff_TEM02=wavelength*1E-7*delta_TEM02_n/2/pi#converts the absorption coefficient(1/cm) to extinction coefficient see Whittaker wavelength converted from nm to cm 
efract_TEM02=ft.hilbert(extinction_coeff_TEM02)#computes the change in refractive index from change in extinction coefficient spectrum
fig,a = plt.subplots(2,sharex=True, figsize=(16, 16)) #sets size of figures
a[0].set_title('TE Reverse Bias Extinction Coefficient Difference Spectra',fontsize=24) #writes title
a[1].set_title('TE Reverse Bias Refractive Index Difference Spectra',fontsize=24) #writes title
a[0].set_ylabel("$\Delta$ Extinction coefficient",fontsize=20)#writes y axis label
a[1].set_ylabel("$\Delta$ Refractive Index",fontsize=20)#writes y axis label
plt.xlabel('Photon energy [eV]',fontsize=20)#writes x axis label
a[0].plot(eV,extinction_coeff_TEM02,'r-',label="Applied Field =37kV/cm") #Plots delta TEM02 extinction coefficient Spectrum
a[0].plot(eV,extinction_coeff_TEM03,'g-',label="Applied Field =74kV/cm") #Plots delta TEM03 extinction coefficient Spectrum
a[0].plot(eV,extinction_coeff_TEM04,'y-',label="Applied Field =111kV/cm") #Plots delta TEM04 extinction coefficient Spectrum
a[0].plot(eV,extinction_coeff_TEM05,'b-',label="Applied Field =148kV/cm") #Plots delta TEM05 extinction coefficient Spectrum
a[1].plot(eV,efract_TEM02,'r-',label="Applied Field =37kV/cm") #Plots delta TEM02 refractive index Spectrum
a[1].plot(eV,efract_TEM03,'g-',label="Applied Field =74kV/cm") #Plots TEM03 Spectrum
a[1].plot(eV,efract_TEM04,'y-',label="Applied Field =111kV/cm") #Plots TEM04 Spectrum
a[1].plot(eV,efract_TEM05,'b-',label="Applied Field =148kV/cm") #Plots TEM05 Spectrum
a[0].legend(loc=0)#writes legend on graph
a[1].legend(loc=0)#writes legend on graph
a[0].grid()#writes grid on graph
a[1].grid()#writes grid on graph
plt.show()
```
